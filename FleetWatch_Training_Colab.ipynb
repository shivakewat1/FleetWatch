## 📊 Understanding the Output

### During Training:
```
-- Episode 1/60 | task: task1-obvious --
  [ENV] Task: task1-obvious
  [LLM] {"anomaly_detected": true, "agent_id": "DRIVER-04", ...}
  [ENV] Server: 0.8077 | Shaping: +0.0800 | Final: 0.8877
  [TRAIN] Loss: 0.0234 | PG: 0.0234 | GradNorm: 0.8234 | Adv: 2.1234
```

**Key Metrics:**
- **Server**: Raw score from API (should vary, not stuck at 0.999)
- **Shaping**: Bonus/penalty applied
- **Final**: Total reward used for learning
- **PG**: Policy gradient loss (should decrease)
- **GradNorm**: Gradient magnitude (should be < 2.0)
- **Adv**: Advantage signal (drives learning)

### Progress Summaries:
```
*** Episodes 1-10 | Avg: 0.7234 | Max: 0.9990 | Min: 0.4567 ***
```

**What to Look For:**
- ✅ Avg increasing within each phase
- ✅ Max reaching 0.85-0.95 on easy tasks
- ✅ Min improving over time

## 🎓 System Details

### Model:
- **Base**: Llama-3-8B-Instruct (4-bit quantized)
- **Method**: LoRA fine-tuning
- **Algorithm**: REINFORCE (simplified, no KL divergence)

### Memory Optimizations:
- Single forward pass per episode
- Aggressive tensor cleanup
- Gradient checkpointing
- MAX_SEQ_LEN: 512 tokens
- MAX_NEW_TOKENS: 128 tokens

### Curriculum:
- **Phase 1 (1-15)**: Task1 only (single agent, obvious)
- **Phase 2 (16-25)**: Task1+2 (patterns)
- **Phase 3 (26-35)**: Task1-3 (adversarial)
- **Phase 4 (36-45)**: Task1-4 (multi-agent cascade)
- **Phase 5 (46-60)**: All 5 (collusion)

### Reward System:
- Multi-agent fuzzy matching
- Partial credit for incomplete answers
- Proportional keyword scoring
- Realistic distribution (0.001-0.999)
